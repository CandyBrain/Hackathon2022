```python
import cv2
import os
import matplotlib.pyplot as plt

image_path = r'D:/Data/mouse_sample_image.jpg' #    .
image = cv2.imread(image_path) #    -> RGB   
    
plt.imshow(image)
plt.show()
```

```python
dx = cv2.Sobel(image, -1, 1, 0, delta=128) # delta     -  0
dy = cv2.Sobel(image, -1, 0, 1, delta=128)

cv2.imshow('src', image)
cv2.imshow('dx', dx)
cv2.imshow('dy', dy)
cv2.waitKey()

# cv2.destroyAllWindows()
```

```python
dx = cv2.Scharr(image, -1, 1, 0, delta=128) # delta     -  0
dy = cv2.Scharr(image, -1, 0, 1, delta=128)

cv2.imshow('src', image)
cv2.imshow('dx', dx)
cv2.imshow('dy', dy)
cv2.waitKey()
```

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt

# reading the image
img = imread('D:/Data/newyork_sample_image.jpg' )
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

```python
resized_img = resize(img, (128*4, 64*4))
plt.axis("off")
plt.imshow(resized_img)
print(resized_img.shape)
```

```python
#creating hog features
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```
