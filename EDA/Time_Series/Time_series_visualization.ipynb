```python
col_list = ['FSR1', 'FSR2', 'FSR3', 'FSR4', 'FSR5', 'FSR6', 'FSR7', 'FSR8', 'FSR9',
       'FSR10', 'FSR11', 'FSR12', 'FSR13', 'FSR14', 'FSR15', 'FSR16', 'FSR17',
       'FSR18', 'FSR19', 'FSR20', 'FSR21', 'FSR22', 'FSR23', 'FSR24', 'FSR25',
       'FSR26', 'FSR27', 'FSR28', 'FSR29', 'FSR30', 'FSR31', 'FSR32', 'FSR33',
       'FSR34', 'FSR35', 'FSR36', 'FSR37', 'FSR38', 'FSR39', 'FSR40', 'FSR41',
       'FSR42', 'FSR43', 'FSR44', 'FSR45', 'FSR46', 'FSR47', 'FSR48', 'FSR49',
       'FSR50', 'FSR51', 'FSR52', 'FSR53', 'FSR54', 'FSR55', 'FSR56', 'FSR57',
       'FSR58', 'FSR59', 'FSR60', 'FSR61', 'FSR62', 'FSR63', 'FSR64']
```

```python
liner_data = pd.read_csv(file_path, names=col_list)
```

```python
fig = plt.figure()
line, = plt.plot([], [], 'bo')
def update(i):
    x = np.linspace(0, len(liner_data))
    line.set_data(x, liner_data.iloc[i])
    plt.clf()
    plt.plot(liner_data.iloc[i].index, liner_data.iloc[i])
    plt.title('Time=%d' % i)

ani = animation.FuncAnimation(fig, update, frames= 175, interval=100, repeat=True)
HTML(ani.to_html5_video())
```
