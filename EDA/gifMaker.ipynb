```python
from matplotlib import pyplot as plt, cm, colors
from IPython.display import HTML
import matplotlib.animation as animation
import numpy as np
import pandas as pd
%matplotlib inline

rawData = pd.read_csv('./subjectTest_1_20220507.csv')
rawData_list = []

for r in rawData.values.tolist():
    rawData_list.append(np.array(r).reshape(8,8))

side = np.linspace(1, 8, 8)
#np.linspace(a, b, n) a b      n .

X, Y = np.meshgrid(side, side)
Z = rawData_list #   (voltage)

fig = plt.figure()
plt.pcolormesh(X, Y, Z[0], shading='auto')
#shading='auto' will decide whether to use 'flat' or 
#'nearest' shading based on the shapes of X, Y and Z.
plt.colorbar() #   
plt.title('Time=%d' % 0) 

def update(i):
    plt.pcolormesh(X, Y, Z[i], shading='auto')
    plt.title('Time=%d' % i)

ani = animation.FuncAnimation(fig, update, frames= 175, interval=100, repeat=True)
# update()  frames      .
HTML(ani.to_html5_video())
```

```python
from matplotlib import pyplot as plt, cm, colors
from IPython.display import HTML
import matplotlib.animation as animation
import numpy as np
import pandas as pd
%matplotlib inline

rawData = pd.read_csv('./subjectTest_1_20220507.csv')
rawData_list = []

for r in rawData.values.tolist():
    rawData_list.append(np.array(r).reshape(8,8))

side = np.linspace(1, 8, 8)
#np.linspace(a, b, n) a b      n .

X, Y = np.meshgrid(side, side)
Z = rawData_list #   (voltage)

fig = plt.figure()
plt.pcolormesh(X, Y, Z[0], shading='nearest') # flat   
#shading='auto' will decide whether to use 'flat' or 
#'nearest' shading based on the shapes of X, Y and Z.
plt.colorbar() #   
plt.title('Time=%d' % 0) 

def update(i):
    plt.pcolormesh(X, Y, Z[i], shading='nearest') # flat   
    plt.title('Time=%d' % i)

ani = animation.FuncAnimation(fig, update, frames= 175, interval=100, repeat=True)
# update()  frames      .
HTML(ani.to_html5_video())

ani.save('seongHoon.gif', writer='imagemagick', fps=70, dpi=100)
```
