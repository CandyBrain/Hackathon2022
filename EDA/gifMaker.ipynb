```python
from matplotlib import pyplot as plt, cm, colors
from IPython.display import HTML
import matplotlib.animation as animation
import numpy as np
import pandas as pd
%matplotlib inline

rawData = pd.read_csv('./subjectTest_1_20220507.csv')
rawData_list = []

for r in rawData.values.tolist():
    rawData_list.append(np.array(r).reshape(8,8))

side = np.linspace(1, 8, 8)
#np.linspace(a, b, n)는 a부터 b 사이를 균일한 간격으로 나누는 숫자 n개를 반환합니다.

X, Y = np.meshgrid(side, side)
Z = rawData_list # 압력센서 값 (voltage)

fig = plt.figure()
plt.pcolormesh(X, Y, Z[0], shading='auto')
#shading='auto' will decide whether to use 'flat' or 
#'nearest' shading based on the shapes of X, Y and Z.
plt.colorbar() # 색상 막대 
plt.title('Time=%d' % 0) 

def update(i):
    plt.pcolormesh(X, Y, Z[i], shading='auto')
    plt.title('Time=%d' % i)

ani = animation.FuncAnimation(fig, update, frames= 175, interval=100, repeat=True)
# update() 함수에 frames에 제공한 값들이 순서대로 제공되어서 프레임을 구성합니다.
HTML(ani.to_html5_video())
```

```python
from matplotlib import pyplot as plt, cm, colors
from IPython.display import HTML
import matplotlib.animation as animation
import numpy as np
import pandas as pd
%matplotlib inline

rawData = pd.read_csv('./subjectTest_1_20220507.csv')
rawData_list = []

for r in rawData.values.tolist():
    rawData_list.append(np.array(r).reshape(8,8))

side = np.linspace(1, 8, 8)
#np.linspace(a, b, n)는 a부터 b 사이를 균일한 간격으로 나누는 숫자 n개를 반환합니다.

X, Y = np.meshgrid(side, side)
Z = rawData_list # 압력센서 값 (voltage)

fig = plt.figure()
plt.pcolormesh(X, Y, Z[0], shading='nearest') # flat 은 오류 뜸
#shading='auto' will decide whether to use 'flat' or 
#'nearest' shading based on the shapes of X, Y and Z.
plt.colorbar() # 색상 막대 
plt.title('Time=%d' % 0) 

def update(i):
    plt.pcolormesh(X, Y, Z[i], shading='nearest') # flat 은 오류 뜸
    plt.title('Time=%d' % i)

ani = animation.FuncAnimation(fig, update, frames= 175, interval=100, repeat=True)
# update() 함수에 frames에 제공한 값들이 순서대로 제공되어서 프레임을 구성합니다.
HTML(ani.to_html5_video())

ani.save('seongHoon.gif', writer='imagemagick', fps=70, dpi=100)
```
